```python
# 2), 3)
matrix_1 = []
for a in range(0,12):
    matrix_1.append([])
    for b in range(0,12):
        matrix_1[a].append(a + b)
    print(matrix_1[a])
            
```

```python
matrix2 = []
for a in range(0,12):
    matrix2.append([])
    for b in range(0,12):
        if a == 0:
            matrix2[a].append(matrix_1[a][b])
        elif b == 0:
            matrix2[a].append(matrix_1[a][b])
        else:
            matrix2[a].append(b + 2*a)
    print(matrix2[a])
```

```python
for a in matrix_1:
    for b in range(0,12):
        if a[b] > 5:
            print(a[b])
```
